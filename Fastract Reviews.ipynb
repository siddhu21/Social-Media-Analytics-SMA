```python
cookie={}
def getAmazonSearch(search_query):
    url="https://www.amazon.in/s?k="+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
    #print(page.content)
```

```python
def Search_1(search_query):
    url="https://www.amazon.in"+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
link=[]
for i in range(len(data_asin)-1):
    response=Search(data_asin[i+1])
#beautifulsoup is a package
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
len(link)
```

```python
r1=[]
for j in range(len(link)):
    for k in range(50):
        response=Search_1(link[j]+'&pageNumber='+str(k))
#beautifulsoup is a package
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            r1.append(i.text)
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser','ner']) #Loads the english model and we dont want prser or an ner so we will disable it
```

```python
reviews=reviews.iloc[0]['reviews']
reviews
```

```python
document = nlp(reviews)#Convert into an Spacy Document
```

```python
print([token for token in document]) ## Prints list of all tokens in the document
```

```python
print([token.text for token in document])
```

```python
stopwords=spacy.lang.en.stop_words.STOP_WORDS
doc_cleaned=[token for token in document if  (token.text not in stopwords)]
print(doc_cleaned)
```

```python
## Let us get POS tags of the words
pos_tags=[(token.text,token.pos_) for token in document if (token.text not in stopwords)]
print(pos_tags)
```

```python
adj=[token for token in document if (token.text not in stopwords) and token.pos_=="ADJ"]
print(adj)
```

```python
doc_cleaned=[token.lemma_ for token in document if (token.text not in stopwords) and token.pos_ not in ['PUNCT','SPACE',"PRON","NUM"]]
print(doc_cleaned)
```
